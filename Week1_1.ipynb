**Week1授業課題1 曾呂利新左衛門問題**

【問題1】100日目の米粒の数

```
%matplotlib inline
import matplotlib.pyplot as plt

list_grain_100 = [2 ** i for i in range(100)]
list_total_grain_100 = []
total = 0
for i in range(100):
    total += 2 ** i
    list_total_grain_100.append(total)
print(list_grain_100)
print(list_total_grain_100)

# 合計で貰った数
print('100日間で合計{:,}個貰った。'.format(list_total_grain_100[-1]))

# グラフを描画
plt.title('number of grains')
plt.xlabel('days')
plt.ylabel('number of grains')
plt.plot(list_grain_100)
plt.show()

plt.title('sum of grains')
plt.xlabel('days')
plt.ylabel('sum of grains')
plt.plot(list_total_grain_100)
plt.show()
```

【問題2】100日目以外の米粒の数

```
# 問題1を関数化する。

def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """

    list_n_grains = [2 ** i for i in range(day)]
    
    list_total_grains = []
    total = 0
    for i in range(day):
        total += 2 ** i
        list_total_grains.append(total)

    return list_n_grains, list_total_grains

during_course_days = 31 + 30 + 31 + 31
list_n_grains, list_total_grains = compute_sorori_shinzaemon(during_course_days)
print('コース受講期間中にもらえる米粒の数は{:,}個'.format(list_total_grains[-1]))


# 先ほどと同様に折れ線グラフを描画してみる
# n日目(n=1, 2, ..., 123)にもらえる米粒の数のグラフ
plt.title('number of grains')
plt.xlabel('day')
plt.ylabel('number of grains')
plt.plot(list_n_grains)
plt.show()

# n日目までの累計のグラフ
plt.title('sum of grains')
plt.xlabel('day')
plt.ylabel('sum of grians')
plt.plot(list_total_grains)
plt.show()
```

【問題3】何人が何日生活できる米か

```
# 引数に米粒の数、人数を入れ、その米でその人数が何日間生活できるかの日数を返す関数

def need_grains(grain, person):
    # お茶碗一杯に入っている米粒はおよそ3250粒。お茶碗一杯150gが普通で、お米は炊くことで重さが2.3倍になることを考慮すると、元々の重さは150/2.3で約65g。65gの米粒を数えると3250粒ほどになるそう。朝、昼、晩と一日3杯とすると1日あたりで一人が必要とする米粒は3250*3となる。
    need_num = 3250 * 3
    survive_days = int(grain / (person * need_num))
    
    return survive_days
  
grain0 = list_total_grains[-1]
survive_days0 = need_grains(grain0, 12)
print('生活できる日数: {:,}日間'.format(survive_days0))
print('生活できる年数: {:,}年間'.format(int(survive_days0 / 365)))
```
